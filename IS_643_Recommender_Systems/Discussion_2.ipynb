## Implementation ideas to prevent recommender systems going bad

We have an ethical obligation not to teach machines or algorithms to be prejudiced. It's not about how smart the data scientists or our algorithms are, but our focus should be entirely on our users and their needs. Some of the issues identified above are powered by some algorithm somewhere. If we leave everything to the algorithms, then it is a problem. Some of the implementation ideas that can mitigate these issues are ->

* Create awareness about data ethics
	- Tell your boss
	- Tell your coworkers
	- Identify groups that could be negatively impacted by your work

* Create interpretable models
    - Only include features that work the way you want
    - Exclude feature interactions that you don't want
    - When a feature like gender is not specified, don't come to conclusions
    
* Diversity controlled testing    
    - To prevent accidental correlation of let's say name / zipcode with race / gender
    
* Data Separation & Ensemble Models
	- Keep interests and features like gender separate
    - Downweight features if you suspect bias or prejudice in the recommendations
